```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
import gradio as gr
from transformers import pipeline
```

```python
model_checkpoint = "t5-base"
revision = "main"
translation_pipeline = pipeline('translation_en_to_de', model=model_checkpoint, revision=revision)
```

```python
def convert_to_lowercase(sentence):
    words = sentence.split()
    first_word = words[0]
    rest_of_words = words[1:]
    converted_words = [first_word[0].upper() + first_word[1:].lower()] + [word.lower() for word in rest_of_words]
    return ' '.join(converted_words)
```

```python
from PIL import Image
def TransSign(inp):
  def convert_to_lowercase(sentence):
    words = sentence.split()
    first_word = words[0]
    rest_of_words = words[1:]
    converted_words = [first_word[0].upper() + first_word[1:].lower()] + [word.lower() for word in rest_of_words]
    return ' '.join(converted_words)
  reader=easyocr.Reader(['en'],gpu=False)
  result=reader.readtext(inp)

  text=""
  for detection in result:
    text=text+detection[1]+" "

  text=convert_to_lowercase(text)
  final_text=translation_pipeline(text)
  return final_text[0]['translation_text']
```

```python
from gradio import Image
iface = gr.Interface(
    fn=TransSign,
    inputs=Image(type="numpy"),
    outputs="text"
)
```

```python
iface.launch()
```
